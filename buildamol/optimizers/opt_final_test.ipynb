```python
import buildamol as bam

mol = bam.read_pdb("/Users/noahhk/GIT/biobuild/docs/_tutorials/ext_final_opt.pdb", "EXT_final")

mol.make_residue_graph().show()
```

```python
g = mol.make_residue_graph()
g.make_detailed(n_samples=0.7)
g.show()
```

```python
g.direct_edges(root_node=mol.get_atom(164), adopt=False)
edges=mol.get_residue_connections()
# edges = g.find_rotatable_edges(min_descendants=30, max_descendants=150)
edges = g.direct_edges(mol.get_atom(164), edges, adopt=False)

v = g.draw()
v.draw_edges(*edges, color="red", elongate=1.2, linewidth=3)
v.show()
```

```python
env = bam.optimizers.DistanceRotatron(g, edges, radius=22, pushback=3)

sol, eval = bam.optimizers.genetic_optimize(env, population_size=75, n_parents=23, n_children=30, n_mutations=12, n_newcomers=10, max_steps=5000, threshold=1e-5)

final = bam.optimizers.apply_solution(sol, env, mol.copy())

final.show()
```

```python
final.to_pdb("final_opt3.pdb")
```

```python
final.make_residue_graph().show()
```

```python
final.find_clashes()
```

```python
g2 = final.make_residue_graph()
g2.make_detailed(n_samples=0.7)
edges2 = g2.find_rotatable_edges(min_descendants=10, max_descendants=30)
edges2 = g2.direct_edges(final.get_atom(164), edges2, adopt=False)
edges2.append(
    final.get_bond(79, 69)
)
v = g2.draw()
v.draw_edges(*edges2, color="red", elongate=1.2, linewidth=3)
v.show()

env = bam.optimizers.DistanceRotatron(g2, edges2, radius=22, pushback=3)

sol, eval = bam.optimizers.genetic_optimize(env)#, population_size=75, n_parents=23, n_children=30, n_mutations=12, n_newcomers=10, max_steps=4000, threshold=1e-5)

final2 = bam.optimizers.apply_solution(sol, env, final.copy())

final2.show()
```
